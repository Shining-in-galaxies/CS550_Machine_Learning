### 1. Voting Classifier

- Import `LogisticRegression`, `DecisionTreeClassifier`, `VotingClassifier` classes
- Import `make_moons` function from `sklearn.datasets`
- Split data into test and train datasets
- Create a voting classifier using logistic regressor and a tree classifier. Use soft voting.
- Print the scores for ensembeled model, logistic regressor and the decision tree.
- If you like, you can show the decision boundaries on a contour plot

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(noise=0.2, random_state=18)
plt.scatter(X[:,0],X[:,1], c=y, cmap='bwr' );
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0, stratify=y)
```

```python
logr = LogisticRegression(C=100)
```

```python
#C*Loss1 + Penalty
```

```python
dt = DecisionTreeClassifier(max_depth=3,random_state=3)
```

```python
vc = VotingClassifier([('logr',logr),('dt',dt)], voting='soft', 
                flatten_transform = False)
```

```python
vc.fit(X_train, y_train)
```

```python
vc_logreg,vc_dt = vc.estimators_
```

```python
print(f'VC score: {vc.score(X_test,y_test):.2f}')
print(f'LogReg score: {vc_logreg.score(X_test,y_test):.2f}')
print(f'DT score: {vc_dt.score(X_test,y_test):.2f}')
```

```python
print(f'VC score: {vc.score(X_test,y_test):.2f}')
print(f'LogReg score: {vc_logreg.score(X_test,y_test):.2f}')
print(f'DT score: {vc_dt.score(X_test,y_test):.2f}')
```
